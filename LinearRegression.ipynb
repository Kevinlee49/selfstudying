```python
pd.concat([freq, mean], axis = 1) # axis 를 안써주면 위 아래로 1열로 붙음, axis = 1 을 써줌으로써 2열로 만들어줌, address 이름은 무시 -> 이웃의 카운트 수, 베이뷰 = 354 건
```

```python
cluster = pd.concat([freq, mean], axis = 1)
cluster = cluster.reset_index()
cluster
```

```python
def get_group(x):
  if x in cluster1['neighborhood']:
      return 'low_price'
  elif x in cluster2['neighborhood']:
      return 'high_price_low_freq'
  else:
      return 'high_price_high_freq'
sf['group'] = sf.neighborhood.apply(get_group)
```

```python
 sf.drop(sf.columns[[0,4,6,7,8,13]], axis=1, inplace=True)
 sf = sf[['bathrooms','bedrooms', 'finishedsqft', 'totalrooms', 'usecode', 'yearbuilt', 'zindexvalue', 'group', 'lastsoldprice']]
```

```python
sf.head()
```

```python
X = sf[['bathrooms','bedrooms', 'finishedsqft', 'totalrooms', 'usecode', 'yearbuilt', 'zindexvalue', 'group']]
Y = sf['lastsoldprice']
```

```python
X
```

```python
X = pd.get_dummies(X, columns = ['usecode', 'group']) # 숫자만 있는 데이터로 바꾸기
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=0) # 랜덤스테잇의 역할 - 항상 같은 결과값을 얻기위해
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
lin_rmse = mean_squared_error(y_pred, y_test) ** 0.5
lin_rmse # 에러의 크기를 나태내주는 지표, 데이터의 스케일에 따라 값이 달라짐, 즉 상대적인 지표, 단 크면 나쁘고 작으면 좋은거, 대신 그 기준은 없음.
```
